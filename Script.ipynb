```python
# Import the data
import pandas as pd
df=pd.read_excel("covid-19-state-level-data.xlsx")
df
```

```python
# Unnamed: 0 is again showing the indices, there is no use of it hence we remove it
del df["Unnamed: 0"]
```

```python
# import the required libraries
import pandas as pd
from datetime import date # to get the date and month
import smtplib as st # to send the emails
months={"month_name":["March","April","May","June"],
       "month_number":[3,4,5,6]}
today=date.today()
month_today=today.month
months_specific_data=df["date"].dt.month==month_today 
index=[]
index_month=months["month_number"].index(month_today)
month_name=months["month_name"][index_month]
for i in months_specific_data.index:
    if months_specific_data[i]==True:
        index.append(i)
monthly_data=df[index[0]:index[-1]] # get the specific month's data
states=monthly_data["state"].unique() # all states
deaths_state_wise=[]
for i in states:
    deaths_state_wise.append(monthly_data[monthly_data["state"]==i]["deaths"].sum())
sorted_deaths_state_wise=deaths_state_wise.copy()
sorted_deaths_state_wise.sort(reverse=True) # arrage the deaths in descending order
index_top1=deaths_state_wise.index(sorted_deaths_state_wise[0])
index_top2=deaths_state_wise.index(sorted_deaths_state_wise[1])
index_top3=deaths_state_wise.index(sorted_deaths_state_wise[2])
state_top1=states[index_top1]
state_top2=states[index_top2]
state_top3=states[index_top3]
percentage_state_top1=round((sorted_deaths_state_wise[0]/monthly_data["deaths"].sum())*100,2)
percentage_state_top2=round((sorted_deaths_state_wise[1]/monthly_data["deaths"].sum())*100,2)
percentage_state_top3=round((sorted_deaths_state_wise[2]/monthly_data["deaths"].sum())*100,2)
# user defined function to get the information
def get_monthly_info():
    return f"Top 3 states with the highest number of covid deaths for the month of {month_name} \n \nMonth - {month_name} \n{state_top1} - {sorted_deaths_state_wise[0]} deaths, {percentage_state_top1}% of total US deaths \n{state_top2} - {sorted_deaths_state_wise[1]} deaths, {percentage_state_top2}% of total US deaths \n{state_top3} - {sorted_deaths_state_wise[2]} deaths, {percentage_state_top3}% of total US deaths"
# to send the email to group users
email_ids=input("Put the email_ids here:").split(" ")
email_ids1=list(map(str,email_ids))
for i in email_ids1:
    s = st.SMTP('smtp.gmail.com', 587)
    s.starttls()
    s.login("desaiomkar767@gmail.com", "mtiyuczfunyuyigb")
    s.sendmail("desaiomkar767@gmail.com",i, get_monthly_info())
    s.quit()
```
